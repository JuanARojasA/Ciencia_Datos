```
import pandas as pd
from sklearn.datasets import load_iris
from sklearn import tree
import graphviz
```

```
df = pd.read_csv('https://raw.githubusercontent.com/JuanARojasA/Ciencia_Datos/master/Taller8/tic-tac-toe.csv', header=None)
df = df.sample(frac=1)
```

```
x = df.iloc[0:958,0:9]
y = df.iloc[0:958,9]
target=['noWin','win']
features=['top-left-square','top-middle-square','top-right-square','middle-left-square',
          'middle-middle-square', 'middle-right-square', 'bottom-left-square',
          'bottom-middle-square', 'bottom-right-square']
```

```
clf = tree.DecisionTreeClassifier()
clf = clf.fit(x, y)
```

```
dot_data = tree.export_graphviz(clf, out_file=None)
graph = graphviz.Source(dot_data)
graph.render("tic-tac-toe")
```

```
dot_data = tree.export_graphviz(clf, out_file=None,
                                feature_names=features,
                                class_names=target,
                                filled=True, rounded=True,
                                special_characters=True)
graph =graphviz.Source(dot_data)
graph
```
